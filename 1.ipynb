```python
#coupled oscillator(2-body)
import numpy as np
import matplotlib.pyplot as plt

# k is spring constant, kapa is interaction constant
t=0
dt=0.01
k=1
kapa=1
m=1
imax=2000 #loop count

# x1=x2, v1=v2  , x1=-x2, v1=-v2 
A=np.array([[(-k-kapa)/m,kapa/m],[kapa/m,(-k-kapa)/m]])
x=np.array([[1],[1]])#[[x1],[x2]]
v=np.array([[1],[1]])#[[v1],[v2]]
dx=np.zeros((2,1))
dv=np.zeros((2,1))

#plot array 
x1=np.zeros((imax,1))
x2=np.zeros((imax,1))
time=np.zeros((imax,1))
t=0

for i in range(imax):
    dv=np.dot(A,x)
    dx=v
    v=v+dv*dt
    x=x+dx*dt
    time[i,0]=i*dt
    x1[i,0]=x[0,0]
    x2[i,0]=x[1,0]  
  
plt.plot(time,x1,label='x1')
plt.plot(time,x2,label='x2',linestyle = '--')
plt.plot(time,x1+x2,label='x1+x2')
plt.xlabel('time',size=15)
plt.ylabel('position',size=15)
plt.legend()
```
